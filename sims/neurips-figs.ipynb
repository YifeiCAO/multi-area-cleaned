```python
# Figure 7a - layer 1 to 2

loadpath = '/Users/michael/Documents/GitHub/multi-area-cleaned/sims/neurips/'
layer2c_store = np.load(loadpath + 'null_potent_dpca_twoarea.npy')

yerr=np.std(layer2c_store, axis=(0)).T / np.sqrt(layer2c_store.shape[0])
x = np.arange(1,120)
y1 = np.mean(layer2c_store, axis=(0)).T[:, 1][1:]
y2 = np.mean(layer2c_store, axis=(0)).T[:, 0][1:]
y3 = np.mean(layer2c_store, axis=(0)).T[:, 2][1:]
yerr1 =yerr[1:, 1]
yerr2 =yerr[1:, 0]
yerr3 =yerr[1:, 2]

ax = plt.subplot(111)

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.plot(x, y1, linewidth=1, color='black')
ax.plot(x, y2, linewidth=1, color='maroon')
ax.plot(x, y3, linewidth=1, color='orange')
ax.fill_between(x, y1-yerr1, y1+yerr1, alpha = 0.5, color='black')
ax.fill_between(x, y2-yerr2, y2+yerr2, alpha = 0.5, color='maroon')
ax.fill_between(x, y3-yerr3, y3+yerr3, alpha = 0.5, color='orange')

plt.xlabel('dimension potent space')
plt.ylabel('Potent projection')
leg = plt.legend(('random','color', 'direction'))
plt.title('2 Area')


for legobj in leg.legendHandles:
    legobj.set_linewidth(4.0)
    
savepath = savepath_figs
plt.savefig(savepath + 'dpca_null_potent_2area_l12' + '.pdf')
```
