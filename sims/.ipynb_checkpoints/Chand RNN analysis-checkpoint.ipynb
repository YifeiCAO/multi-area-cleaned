```python
ntrials = 2800
nunits = 300
ntime = 260
rates = np.zeros((nunits, ntrials, ntime))
for i in range(ntrials):
    rates[:, i, :] = psth.trials[i]['r'][:, :ntime]    
zrates = rates

# compute the units zscore across trial and time
m = np.mean(zrates, axis=(1, 2))
std = np.std(zrates, axis=(1, 2))
zrates2 = (zrates.reshape(nunits, ntrials * ntime) - m.reshape(nunits, 1)).reshape(nunits, ntrials, ntime)
zrates3 = (zrates2.reshape(nunits, ntrials * ntime) / std.reshape(nunits, 1)).reshape(nunits, ntrials, ntime)
```

```python
# psth.set_align(align='cb')
# psth.sort = ['cohs', 'cols']
# psth.gen_psth()
from __future__ import division
def coh(coh):
    return 2 * coh/225 - 1

ntrials = 2800
ncoef = 3
F = np.zeros((ncoef, ntrials))
# F[2, :] = 1

for i in range(ntrials):
    F[0, i] = psth.trials[i]['info']['choice']
    F[1, i] = coh(psth.trials[i]['info']['cond']) # maybe divide by 0.9022 so that it's normalized
    F[2, i] = psth.trials[i]['info']['left_right']
    
A = np.linalg.inv(F.dot(F.T)).dot(F)
betas = np.tensordot(A, np.swapaxes(zrates3, 0, 1), axes=([1], [0]))
```

```python
def multivariate_normal(x, d, mean, covariance):
    x_m = x - mean
    # eps = 1e-50
    # pdb.set_trace()
    return (1. / ((np.sqrt((2 * np.pi)**d * np.linalg.det(covariance)))) * np.exp(-(np.linalg.solve(covariance, x_m).T.dot(x_m)) / 2))
    # return (np.exp(-(np.linalg.solve(covariance, x_m).T.dot(x_m)) / 2))


def calculate_kl(z1, z2, num_samples=1000, units=100):
    m1 = np.mean(z1, axis=1)  # sample mean
    cov1 = np.cov(z1)  # sample covariance
    m2 = np.mean(z2, axis=1)
    cov2 = np.cov(z2)
    # pdb.set_trace()

    x_pk = np.random.multivariate_normal(m1, cov1, num_samples)  # draw samples from one distribution
    px = np.zeros((num_samples, 1))
    qx = np.zeros((num_samples, 1))

    # there may and probably is a better way to vectorize this
    for i in range(num_samples):
        px[i] = multivariate_normal(x_pk[i, :], units, m1, cov1)  # compute the probability that the samples came from px
        qx[i] = multivariate_normal(x_pk[i, :], units, m2, cov2)  # compute the probability that the samples came from qx

    # print("{} is pk.shape".format(pk.shape))
    kl = np.mean(np.log2(2 * px) - (np.log2(qx + px)))  # MI in bits
    # kl = np.mean(np.log2(px) - np.log2(qx))
    return kl
```

```python
kl_1_store = np.zeros((261, ))
kl_2_store = np.zeros((261, ))
kl_3_store = np.zeros((261, ))

for i in np.arange(50, 261):
    print(i)
    rates1 = m[:, :, i]
    rates2 = n[:, :, i]

    idx4 = np.arange(0, 7)
    idx5 = np.arange(70, 75)
    idx6 = np.arange(288, 293)

    kl_1 = calculate_kl(rates1[idx4, :], rates2[idx4, :], 1000)
    kl_2 = calculate_kl(rates1[idx5, :], rates2[idx5, :], 1000)
    kl_3 = calculate_kl(rates1[idx6, :], rates2[idx6, :], 1000)

    kl_1_store[i] = kl_1
    kl_2_store[i] = kl_2
    kl_3_store[i] = kl_3
```
